```python
model = YOLO('/content/gdrive/MyDrive/temp/hw8/runs/detect/train2/weights/best.pt')
```

```python
TEST_PATH = '/content/gdrive/MyDrive/temp/hw8/test'
fnames = [os.path.join(TEST_PATH, f) for f in os.listdir(TEST_PATH) if os.path.isfile(os.path.join(TEST_PATH, f))]
```

```python
img = Image.open(fnames[0]).convert('RGB')
res = model(img)
Image.fromarray(res[0].plot()[:, :, ::-1])
```

```python
from tqdm import tqdm
import os
```

```python
save_dir = '/content/gdrive/MyDrive/temp/hw8/preds'
```

```python
for fn in tqdm(fnames):
    img = Image.open(fn).convert('RGB')
    trg = '.'.join(fn.split('/')[-1].split('.')[:-1]) + '_pred.jpg'
    res = model(img)
    pred_img = Image.fromarray(res[0].plot()[:, :, ::-1])
    pred_img.save(os.path.join(save_dir, trg))
```
